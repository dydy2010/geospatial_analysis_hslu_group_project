```python
import pandas as pd
import os
```

### Traffic dataset

```python
data_path = "../data/raw"

files = [
    f for f in os.listdir(data_path)
    if f.endswith(".csv") and f.startswith("sid_dav")
]

dfs = []
for file in files:
    df = pd.read_csv(os.path.join(data_path, file))
    dfs.append(df)

files 
```

```python
df_all = pd.concat(dfs, ignore_index=True)
df_all.head(2)
```

```python
df_all.shape
```

```python
#df_all["MessungDatZeit"] = pd.to_datetime(df_all["MessungDatZeit"])
```

```python
# Drop unneccesary columns
df_all = df_all.drop(columns=["HNr", "D1ID", "D2ID", "D3ID", "D4ID", "LieferDat"])
df_all.head()
```

```python
# Translate column headers into English
df_all = df_all.rename(columns={
    "MSID": "measurement_site_id",
    "MSName": "measurement_site_name",
    "ZSID": "counting_site_id",
    "ZSName": "counting_site_name",
    "Achse": "axis",
    "HNr": "house_number",
    "Hoehe": "position_description",
    "EKoord": "east_coordinate",
    "NKoord": "north_coordinate",
    "Richtung": "direction",
    "Knummer": "signal_id",
    "Kname": "signal_name",
    "AnzDetektoren": "num_detectors",
    "MessungDatZeit": "timestamp",
    "AnzFahrzeuge": "vehicle_count",
    "AnzFahrzeugeStatus": "vehicle_count_status"
})
df_all.head(2)
```

```python
# Replace values in measurement_site_name and position_description, translate into English
df_all["measurement_site_name"] = df_all["measurement_site_name"].replace({
    "Unbekannt": "Unknown"
})

df_all["position_description"] = df_all["position_description"].replace({
    "Unbekannt": "Unknown"
})

# Replace direction values
df_all["direction"] = df_all["direction"].replace({
    "einwärts": "inbound",
    "auswärts": "outbound",
    "einwärts (ohne Busse)": "inbound (excluding buses)"
})

# Replace vehicle_count_status values
df_all["vehicle_count_status"] = df_all["vehicle_count_status"].replace({
    "Gemessen": "Measured",
    "Fehlend": "Missing",
    "Imputiert": "Imputed"
})

df_all.head(2)
```

```python
# Remove content inside brackets
df_all["counting_site_name"] = df_all["counting_site_name"].str.replace(
    r"\s*\(.*?\)", "", regex=True
)
```

```python
# Harmonize "Autobahn" entries to, e.g. "A1*
df_all["counting_site_name"] = df_all["counting_site_name"].str.replace(
    r"Autobahn\s+([A-Za-z0-9]+)\s*-.*", r"\1", regex=True
)
```

```python
df_all.head(2)
```

```python
output_path = "../data/processed"
df_all.to_csv(os.path.join(output_path, "traffic_data_cleaned.csv"), index=False)
```

```python
# Transate columns headers into English
df_pop = df_pop.rename(columns={
    "StichtagDat": "reference_date_year",
    "SexCd": "sex_code",
    "SexLang": "sex",
    "HerkunftCd": "origin_code",
    "HerkunftLang": "origin",
    "KreisCd": "city_district",
    "QuarLang": "statistical_quarter",
    "AnzBestWir": "population_count"
})
```
